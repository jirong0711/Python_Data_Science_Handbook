```python
import numpy as np
x = np.linspace(0,10,100)

fig = plt.figure()
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")
```

```python
#matplotlib가 보이는 두가지 인터페이스 보기
    #매트랩 스타일 인터페이스
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
    ##객체지향 인터페이스
fig, ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
#1번째 방식 축따라 그리기
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x)) #2번째 방식
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))
plt.xlim(10,0) #축순서 뒤바꾸기
plt.ylim(1.5,-1.5)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x,np.sin(x), '--g', label="sin(x)")
plt.plot(x,np.cos(x),":b", label="cos(x)")
plt.axis("equal")

plt.legend()
```

```python
#객체지향 메서드 방식
ax = plt.axes()
ax.plot(x,np.sin(x))

ax.set(xlim = (0,10), ylim = (-2,2), xlabel="x", ylabel="sin(x)", title="A Simple Plot")
```
